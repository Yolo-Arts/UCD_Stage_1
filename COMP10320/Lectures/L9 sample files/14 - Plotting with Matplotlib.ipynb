# Plotting with Matplotlib

Matplotlib is the standard Python plotting library. It provides a variety of plots and visualisations: 

https://matplotlib.org/stable/index.html

To use it in an Jupyter notebook, we firstly set up our notebook as follows:

```python
import matplotlib
import matplotlib.pyplot as plt
# important! we need to tell Jupyter to display plots within the notebook
%matplotlib inline
```

## Matplotlib Basics

Matplotlib visualisations consists of a figure, which contains one or more plots. A simple plot can be used to show the values in a standard Python list. Here the numeric values are plotted on the y-axis and the days will be plotted on the x-axis.

```python
# this is the data that we want to plot
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
btc = [7875.1,  7908.7, 7665.3,  7386.4,  6137.5,  6383.5, 6310.1]
```

```python
# create a new figure, setting the dimensions of the plot
plt.figure(figsize=(8, 4.5))
# set up the plot
plt.plot(days, btc);
```

Note: the ; in the above code supresses the Notebook output. Try removing it an see what happens. 

We can customise this plot by changing its dimensions, changing the line colour, adding a title, and adding text labels to the axes:

```python
# create a new figure, setting the dimensions of the plot
plt.figure(figsize=(8, 4.5))
# set the line colour and increase the line thickness
plt.plot(days, btc, color="coral", linewidth=3)
# set the plot title, with the specified font size
plt.title('Value of Bitcoin', fontsize=13)
# set the range on the y-axis
plt.ylim(0,10000)
# add the y-grid lines
plt.grid(axis="y")
# set the axis labels, with the specified font size
plt.xlabel('Day', fontsize=13)
plt.ylabel('Value ($)', fontsize=13)
plt.show()
```

We can also plot multiple lines on the same plot:

```python
btc = [7875.1, 7908.7, 7665.3, 7386.4, 6137.5, 6383.5, 6310.1]
rbtc = [1189.8, 1196.0, 1163.3, 1189.0, 2201.3, 2189.0, 2100.1]
etr = [5668.2, 5908.6, 5568.3, 5286.3, 5056.8, 4883.8, 4720.2]
```

```python
# create a new figure, setting the dimensions of the plot
plt.figure(figsize=(8, 4.5))
# set the line colour and increase the line thickness
plt.plot(days, btc, color="coral", linewidth=3)
plt.plot(days, rbtc, color="green", linewidth=3)
plt.plot(days, etr, color="blue", linewidth=3)
# set the plot title, with the specified font size
plt.title('Value of Cryptocurrencies', fontsize=13)
# set the range on the y-axis
plt.ylim(0,10000)
# add the y-grid lines
plt.grid(axis="y")
# set the axis labels, with the specified font size
plt.xlabel('Day', fontsize=13)
plt.ylabel('Value ($)', fontsize=13)
# add a legend
plt.legend(["BTC","RBTC","ETR"], fontsize=12)
plt.show()
```

##  Scatter Plots
Next let's create a scatter plot, using two Python lists of identical size for the X and Y values. 

For further details on Matplotlib scatter plots, see: https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.scatter.html

```python
# this is our data
x = [1, 5, 8, 3, 7]
y = [0.4, 0.25, 0.65, 0.7, 0.2]
# create the figure
plt.figure(figsize=(6, 5))
# create a scatter plot, specifying the size of the points
plt.scatter(x, y, s=100)
# customise the range of the axes: values are [xmin xmax ymin ymax]
plt.axis([0,10,0,1])
plt.show()
```

Let's try another scatter plot, based on randomly-generated values:

```python
# Generate the random values
import random    
r_x_data = []
r_y_data = []
for i in range(10):    
    # generate a random x value between 0 and 1
    r_x_data.append(random.random())
    # generate a random y value, between 0 and 10
    r_y_data.append(random.random() * 10)
```

```python
r_x_data
```

```python
r_y_data
```

We create a new figure, then pass the X and Y values to the function *plt.scatter()*:

```python
# set up the new figure
plt.figure(figsize=(6, 5))
# set up the scatter plot
p = plt.scatter(r_x_data, r_y_data, s=200)
plt.show()
```

Let's regenerate the plot, this time cutomising the appearance of the scatter plot:

```python
# change the size of the figure
plt.figure(figsize=(7, 5.5))
# we will change the colour, shape, and size for each point
p = plt.scatter(r_x_data, r_y_data, c="green", s=150, marker="x")
# set the axis labels
plt.xlabel("X-Values", fontsize=13)
plt.ylabel("Y-Values", fontsize=13)
# add a grid
plt.grid()
# increase the font sizes on the axes
plt.xticks(fontsize=13)
plt.yticks(fontsize=13);
plt.show()
```

We can show multiple plots on the same figure:

```python
# set up the new figure
plt.figure(figsize=(7, 5.5))
# generate and plot 4 groups of points
for color in ["red","green","blue","orange"]:
    # generate 8 random points
    x = []
    y = []
    for i in range(8):
        x.append(random.random())
        y.append(random.random())
    # show the plot for this data, using a different colour each time
    plt.scatter(x, y, c=color, s=75, label=color)
# increase the font sizes on the axes
plt.xticks(fontsize=13)
plt.yticks(fontsize=13);
plt.show()
```

## Bar Charts

Next we will look at creating vertical **bar charts**. 

See: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html

```python
values = [5, 11, 14, 6]
names = ["Alice", "Paul", "Susan", "Bob"]
```

```python
# these are the corresponding positions on the y-axis, for plotting purposes
y_pos = range(len(names))
y_pos
```

Now we can display this as a vertical bar chart:

```python
plt.figure(figsize=(8, 5))
# set up the bar chart
plt.bar(names, values, color="purple", zorder=3)
plt.title("Purchases by Customer", fontsize=13)
plt.ylabel("Number of Purchases", fontsize=13)
# add a grid for the y-axis
plt.grid(axis='y')
# increase the font sizes on the axes
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

Alternatively, we could create a horizontal bar chart for the same data using the *barh()* function:

```python
plt.figure(figsize=(8, 5))
# set up the bar chart
plt.barh(y_pos, values, align='center', color="green", zorder=3)
plt.title("Purchases by Customer", fontsize=13)
# specify the y-axis labels and their position
plt.yticks(y_pos, names, fontsize=13)
plt.xticks(fontsize=13)
plt.xlabel("Number of Purchases", fontsize=13)
# add a grid for the x-axis
plt.grid(axis='x');
plt.show()
```

## Pie Charts
Matplotlib can also produce pie charts. See: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

```python
# some sample data
counts = [33, 28, 18]
categories = ["Under 30", "30 to 49", "50+"]
```

```python
# set up the new figure, specifying the size of the figure
plt.figure(figsize=(6, 6))
# create the pie chart on the sample data
p = plt.pie(counts, labels=categories, colors=["darkorange","darkgreen","purple"])
plt.show()
```

We will generate a second pie chart, where we further customise some of the aspects of the chart. We'll then save the figure to disk as a PNG file.

```python
# set up the new figure, specifying the size of the figure
plt.figure(figsize=(6, 6))
# create the pie chart on the sample data
p = plt.pie(counts, labels=categories, colors=["darkorange","darkgreen","purple"], 
            autopct="%.1f%%", shadow=True, startangle=90)
plt.show()
# save the figure
plt.savefig("sample-pie-chart.png")
```

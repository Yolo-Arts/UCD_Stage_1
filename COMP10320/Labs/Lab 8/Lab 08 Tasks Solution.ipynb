# Lab 7 Tasks - Solution

In this notebook we will apply **data visualisation** methods in Pandas to a cleaned version of a dataset from an Irish triathlon. In the dataset, each row represents an athlete, described by a number of different descriptive features:

- *Number:* The athlete's race bib number
- *Place:* The place in which the athlete finished the race
- *Age:* The athlete's age
- *AgeCategory:* A set of categories representing athlete ages
- *Gender:* The gender that the athlete declared ('M' or 'F')
- *Province:* The Irish province where the athlete comes from (Leinster, Munster, Connacht, Ulster)
- *Swim:* The time taken for the swimming segment of the event (in seconds)
- *T1:* The time taken for the first transition of the event, from cycling to swimming (in seconds)
- *Cycle:* The time taken for the cycling segment of the event (in seconds)
- *T2:* The time taken for the swimming segment of the event, from swimming to running (in seconds)
- *Run:* The time taken for the running segment of the event (in seconds)
- *Finish:* The total time taken for the race by an athlete

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

## Task 1 - Data Loading

This exercise uses the file triathlon_clear.csv, which is available on BrightSpace.

Begin by loading the dataset into a Pandas DataFrame, where the row index will be given by the athlete's bib number. 

Display the first 10 rows of the DataFrame.

```python
# we could either download the file using urllib, or we can use Pandas to do this for us
df = pd.read_csv("triathlon_clean.csv", index_col="Number")
```

## Task 2 - Data Visualisation

Use **bar charts** to visualise:
1. The number of athletes per age category
2. The number of athletes per province

(Hint: Start by calculating a frequency table for the relevant column)

```python
# get the frequency counts
age_counts = df["AgeCategory"].value_counts()
# plot the frequencies
ax = age_counts.plot(kind="bar", fontsize=13, color="navy", figsize=(6, 4.5))
plt.ylabel("Number of Athletes", fontsize=13);
plt.show()
```

```python
# get the frequency counts
province_counts = df["Province"].value_counts()
# plot the frequencies
ax = province_counts.plot(kind="bar", fontsize=13, color="teal", figsize=(6, 4.5))
plt.ylabel("Number of Athletes", fontsize=13);
plt.show()
```

Proudce a **pie chart** showing the percentage of athletes by province

(Hint: Start by calculating a frequency table for the relevant column)

```python
# get the frequency counts
province_counts = df["Province"].value_counts()
# plot the frequencies
ax = province_counts.plot(kind="pie", figsize=(5, 5), fontsize=12, autopct="%.1f%%", 
                        colors=["royalblue", "red", "coral", "darkgreen"])
ax.set_ylabel(None);
plt.show()
```

Produce a **histogram** visualisation of the distribution of athlete finish times:

```python
ax = df["Finish"].hist(bins=10, figsize=(8, 4.5), color='green', grid=False, rwidth=0.9)
plt.ylabel("Number of Athletes", fontsize=13)
plt.xlabel("Finish Time (seconds)", fontsize=13);
plt.show()
```

Repeat the above, but this time produce a **histogram** visualisation of the distribution of finish times for female athletes only:

```python
# filter based on gender first
df2 = df[df["Gender"]=="F"]
# then generate the histogram
ax = df2["Finish"].hist(bins=10, figsize=(8, 4.5), color='green', grid=False, rwidth=0.9)
plt.ylabel("Number of Athletes", fontsize=13)
plt.xlabel("Finish Time (seconds)", fontsize=13);
plt.show()
```

Produce three **scatter plots** which show: 
1. The relationship between the time taken for the swimming and cycling segments.
2. The relationship between the time taken for the swimming and running segments.
3. The relationship between the time taken for the cycling and running segments.

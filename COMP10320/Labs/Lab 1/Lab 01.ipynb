## Question 1

Write a program that takes as input a temperature in degrees Celsius and converts it to a temperature in degrees Fahrenheit.

The formula for converting Celsius to Fahrenheit is: F = C*9/5 + 32

The user should be prompted with the message: "Please enter a temperature in degrees Celsius:" 

The computer should reply: "The temperature in degrees Fahrenheit is F."

```python
# Add your code here.

celsius = float(input("Please enter a temperature in degrees Celsius"))

Fahrenheit = celsius * 9 / 5 + 32
print(f"The temperature in degrees Fahrenheit is {Fahrenheit}")
```

```python
# Add your code here.
length = float(input("Enter a length"))

areaOfSquare = length * length
volumeOfCube = length * length * length
areaOfCircle = 3.14 * length * length

print(f"The area of a square with side of that length is {areaOfSquare}")
print(f"The volume of a cube with side of that length is {volumeOfCube}")
print(f"The area of a circle with side of that length is {areaOfCircle}")
```

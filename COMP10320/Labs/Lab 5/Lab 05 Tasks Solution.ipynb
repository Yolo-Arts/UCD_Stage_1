## Task 1 - JSON Parsing

Use Python to download a file containing metadata for member states of the EU in JSON format from the URL: 

http://mlg.ucd.ie/modules/COMP30760/eu.json

```python
import urllib.request
url = "http://mlg.ucd.ie/modules/COMP30760/eu.json"
response = urllib.request.urlopen(url)
raw_json = response.read().decode()
print(raw_json)
```

Parse the JSON data that you have downloaded.

```python
import json
eu_data = json.loads(raw_json)
print(eu_data)
```

From the parsed JSON data, print a list of the all of the country names, along with the corresponding capital city.

```python
countries = eu_data["countries"]
for country in countries:
    print("%s = %s" % ( country["name"], country["capital_city"] ) )
```

Extract the population information for each member state.

```python
for country in countries:
    print( "%s = %s" % ( country["name"], country["population"] ) )
```

Export the EU member state data to a new file in CSV format, with an appropriate header line.

```python
import csv
# open the output file for writing
fout = open("eu.csv", "w")
# specify the ordered list of fields in our movie data
fields = ["name", "population", "capital_city", "first_language"]
writer = csv.DictWriter(fout, fieldnames=fields)
# write the header row
writer.writeheader()
# write each row of data
for country in countries:
    writer.writerow(country)
fout.close()
```

```python
# we could check the CSV output
for line in open("eu.csv","r"):
    print(line.strip())
```

## Task 2 - XML Parsing

Use Python to download a file containing a contact list in XML format from the URL: 

http://mlg.ucd.ie/modules/COMP30760/contacts.xml

```python
url = "http://mlg.ucd.ie/modules/COMP30760/contacts.xml"
response = urllib.request.urlopen(url)
raw_xml = response.read().decode()
print(raw_xml)
```

Parse the XML data that you have downloaded.

```python
import xml.etree.ElementTree
tree = xml.etree.ElementTree.fromstring(raw_xml)
```

From the parsed XML data, extract the name, email address and phone number of each contact. 

Store these contacts in a list of dictionaries, and print them out.

```python
contacts = []
for entry in tree.findall("contact"):
    contact = {}
    contact["name"] = entry.find("name").text
    contact["email"] = entry.find("email").text
    contact["phone"] = entry.find("phone").text
    contacts.append(contact)
print("Parsed %d contacts" % len(contacts))
```

```python
for contact in contacts:
    print("%s:\temail=%s\tphone=%s" % ( contact["name"], contact["email"], contact["phone"] ) )
```

Export the contact data to a new file in CSV format.

```python
import csv
# open the output file for writing
fout = open("contacts.csv", "w")
# specify the ordered list of fields in our data
fields = ["name", "email", "phone"]
writer = csv.DictWriter(fout, fieldnames=fields)
# write the header row
writer.writeheader()
# write each row of data
for contact in contacts:
    writer.writerow(contact)
fout.close()
```

Use Pickle to serialise the contact data to a file. Verify that you can deseralise the data again.

```python
import pickle
# use Pickle to write the contact data out to a file
fout = open("contacts.pkl", "wb")
pickle.dump(contacts, fout)
fout.close()
```

```python
# verify that this worked by reading it back in again
fin = open("contacts.pkl", "rb")
backup = pickle.load(fin)
fin.close()
print("Read %d contacts" % len(backup))
for contact in backup:
    print("%s:\temail=%s\tphone=%s" % ( contact["name"], contact["email"], contact["phone"] ) )
```

## Task 3 - Web Scraping and Data Extraction

For this task, we will use the *Beautiful Soup* package version 4:

https://beautiful-soup-4.readthedocs.io/

If the package is not already installed on your machine, you will need to install this via *pip*:

> pip install beautifulsoup4

The EU member state data we saw in Task 1 is also available in the following HTML page:

http://mlg.ucd.ie/modules/COMP30760/eu.html

Use Python to download the file and use the BeautifulSoup package to extract information about EU member states from the HTML.

```python
# first retrieve the HTML page
import urllib
url = "http://mlg.ucd.ie/modules/COMP30760/eu.html"
response = urllib.request.urlopen(url)
html = response.read().decode()
```

```python
# next we will use BeautifulSoup
import bs4
# parse the HTML content
soup = bs4.BeautifulSoup(html, 'html.parser')

# find all table rows which contain the information - i.e. <tr> tags
rows = soup.find_all('tr')

countries_data = []
# process each row
for row in rows:
    # find all columns in the row - i.e. <td> tags
    columns = row.find_all('td')
    # if there are no <td> tags, this is the header, so skip it
    if len(columns) == 0:
        continue
    country_data = {
        "Country": columns[0].text.strip(),
        "Population": columns[1].text.strip(),
        "First Language": columns[2].text.strip(),
        "Capital City": columns[3].text.strip()
    }
    # we could do some extra cleaning on the population value to turn it into a number
    # note this requires removing the commas
    country_data["Population"] = int(country_data["Population"].replace(",",""))
    countries_data.append(country_data)
```

```python
# display the data that we have found
for country in countries_data:
    print(country)
```

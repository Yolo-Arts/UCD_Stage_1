## Task 1

Write a new Python function which reads in two text files containing numbers (numbers1.txt and numbers2.txt), and finds the numbers which are overlapping (i.e. the intersection of the numbers present in both files). The output should take the form:<br><br>

Overlap is \{....}<br>

```python
# Define a new function to read a set of numbers from the specified file
def read_numbers( fileName ):
	numbers = set()
	fin = open(fileName,"r")
	for l in fin.readlines():
		l = l.strip()
		num = int(l)
		numbers.add( num )
	fin.close()
	return numbers

# Read the first file
numbers1 = read_numbers("numbers1.txt")

# Read the second file
numbers2 = read_numbers("numbers2.txt")

# Get the overlap (i.e. the intersection between the two sets)
overlap = numbers1.intersection(numbers2)
print( "Overlap is %s" % overlap )
```

## Task 2

Write a Python function which reads in the text file names.txt, line by line, and displays only the specified lines. The lines to be read should be specified by the user with ',' separated integers at a prompt. For example:

Choose lines to display:<br>
You chose lines: ???<br>
Name 1 = Michael<br>
Name 3 = Kaylee<br>
Name 12 = Emily<br>

Note: There is no need to implement error handling in this task; you will do that in Task 3. You can assume that the user will always enter ',' separated integers at the prompt. 

```python
# Find out what line numbers we want
userInput = input('Choose lines to display:')
print("You chose the lines: " + userInput)

# Covert userInput into a list of strings, split on ","
lines = userInput.split(",")

line_numbers = []
# Convert the line numbers to integers
for s in lines:
    line_numbers.append(int(s))

# Now read the file
fin = open("names.txt","r")

current_line_number = 0

for l in fin.readlines():
    # increment the line number
    current_line_number += 1
    # Do we want to print this one?
    if current_line_number in line_numbers:
        print("Name %d = %s" % (current_line_number, l.strip() ))

fin.close()
```

## Task 3

Modify the function in Task 2 by adding a user prompt to specify a file to open. Also, add exception handling for the cases where: (1) the specified text file does not exist, (2) an invalid number is specified. For example:

Choose a file to open:
You chose the file: ???
Choose lines to display:<br>
You chose the lines: ???
Name 1 = Michael<br>
Name 3 = Kaylee<br>
Name 12 = Emily<br>

```python
try:
    # Find out what file we want
    fileName = input('Choose file to open:')
    print("You chose the file: " + fileName)

    # Now read the file
    fin = open(fileName,"r")

    # Find out what line numbers we want
    userInput = input('Choose lines to display:')
    print("You chose the lines: " + userInput)

    # Covert userInput into a list of strings, split on ","
    lines = userInput.split(",")

    line_numbers = []
    for s in lines:
        line_numbers.append(int(s))

    current_line_number = 0
    for l in fin.readlines():
        # increment the line number
        current_line_number += 1
        # Do we want to print this one?
        if current_line_number in line_numbers:
            print("Name %d = %s" % (current_line_number, l.strip() ))

    fin.close()

except ValueError as e:
    print("Invalid line number specified: %s. Ending script" % e )
    
except IOError as e:
    print("Unable to read from file: %s. Ending script" % e )
```

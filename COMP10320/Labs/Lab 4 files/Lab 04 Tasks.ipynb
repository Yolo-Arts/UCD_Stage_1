## Task 1

Write a new Python function which reads in two text files containing numbers (numbers1.txt and numbers2.txt), and finds the numbers which are overlapping (i.e. the intersection of the numbers present in both files). The output should take the form:<br>

The overlap is ... <br>

```python
try:
    with open("numbers1.txt", "r") as file_numbers1:
        numbers1 = file_numbers1.readlines()
        for i in range(len(numbers1)):
            numbers1[i] = numbers1[i].strip()
except:
    print("File could not be opened")

try:
    with open("numbers2.txt", "r") as file_numbers2:
        numbers2 = file_numbers2.readlines()
        for i in range(len(numbers2)):
            numbers2[i] = numbers2[i].strip()
except:
    print("File could not be opened")

print(numbers1)
print(numbers2)

overlap = []
for i in range(len(numbers1)):
    for j in range(len(numbers2)):
        if numbers1[i] == numbers2[j]:
            overlap.append(numbers1[i])

print(f"The overlap is: {overlap}")
```

## Task 4

Write a Python script which reads in the text file names.txt, line by line, and displays a random sample of N names, where N is a user-specified number. Your script should handle the cases where: (1) the specified file does not exist, (2) an invalid number N is specified.  For example:

Choose lines to display:<br>
3<br>
Name 1 = Ethan<br>
Name 2 = Lucas<br>
Name 3 = Caden<br>

```python
import random

try:
    with open("names.txt", "r") as file_names:
        names = file_names.read().splitlines()
except:
    print("File could not be opened")

# print(len(names))
# print(names[16])

num = int(input("Choose lines to display: "))
while num > 0 and num > len(names) - 1:
    print("Invalid input")
    num = int(input("Choose lines to display: "))

for i in range(num):
    print(f"Name {i} = {random.choice(names)}")
```
